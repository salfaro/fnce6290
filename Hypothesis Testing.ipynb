```python
#Test the claim that 75% of US adults have smartphones at the 5% level of significance. 
#A sample of 900 adults showed that 77% had smartphones.  

#We can use the normal distribution because the sample size is large and n*(1-p)= 225 > 10.  
#This is a two-tailed test so the critical z-values will be ±1.96.   
#The Z-score would be (0.77 - 0.75)/√(0.25*0.75/900) = 0.02/0.0144 = 1.385, 
#which is between the lower and upper critical values of ±1.96 so is in the acceptance region.  

ZScore = (0.77-0.75) / math.sqrt(0.25*0.75/900)

print(ZScore)

#We cannot reject H0.
```

```python
#Example: We will test this claim. The standard deviation of returns to Apple stock have increased.  The historical standard deviation (20170103 through 20181130) is 1.425%. We will sample data for 19 trading days in December 2018.
#
#Step 1: Create our hypotheses.  This is a one-tailed test since the statement states strictly 'increased'.
#
#H0: σ ≤ 1.425%
#H1: σ > 1.425%
#The rejection region will be in the right-hand tail of the distribution. 
#
#Step 2: Set a level of significance.  Typically this will be 10%, 5% or 1%.  Lets use 5%.
#
#Step 3: Find the critical χ2 values for (19-1 = 18) degrees of freedom. We use the Excel function =CHISQ.INV(95%,18) = 28.8693.
#
#Step 4:  Collect our sample and compute the sample standard deviation, s. We do this and find s=2.809%.
#
#Step 5: Compute our test statistics: 18*2.8092 /1.4252 =69.9378.
#
#Step 6: Make a decisions.  Since the test statistic of 69.94 is much larger than the critical value of 28.87, we reject the null hypothesis and conclude that the standard deviation is different than 1.425%.
```

```python
#We can find the critical values for a chi-squared distribution using the Excel function, 
#CHISQ.INV(probability,df).  If we want the value that separates the left-most 5% 
#of the area under the curve with 10 degrees of freedom we would enter: =CHISQ.INV(5%,10) = 3.94. 
#The value that delineates the 5% area in the right tail would be =CHISQ.INV(95%,10) = 18.307.
print(stats.chi2.ppf(0.05, 10))
print(stats.chi2.ppf(0.95, 10))
```
